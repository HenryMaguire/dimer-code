```python
def calculate_steadystate_comparison(PARAMS, num, threshold=0.):
    T_values = np.linspace(30, 300, 60)
    biases = np.array([ 0*1e-3, 10*1e-3, 30*1e-3])*ev_to_inv_cm
    ti =time.time()
    rho_nonadd_list = []
    op_list = []
    for bias in biases:
        PARAMS = PARAMS_update_bias(PARAMS, bias_value=bias)
        rho_nonadd = []
        opss = []
        for j, T in enumerate(T_values):
            PARAMS.update({'T_1':T, 'T_2': T})
            
            H, L, PARAMS = get_H_and_L_full(PARAMS,silent=True, threshold=1e-8)

            rho_nonadd.append(steadystate(H[1], [L['nonadd']]))

            opss.append(make_expectation_operators(PARAMS))
            if j in [1,10,20,30,40,49]:
                print(bias)
        rho_nonadd_list.append(rho_nonadd)
        op_list.append(opss)
    data_dict = {}
    filename = 'DATA/T_dependence/3ls_{}'.format(num)
    PARAM_names = ['rho_nonadd_list', 'op_list', 'T_values', 'PARAMS', 'biases']
    scope = locals() # Lets eval below use local variables, not global
    data_dict.update(dict((name, eval(name, scope)) for name in PARAM_names))
    print("Calculation took {} seconds".format(time.time()-ti))
    try:
        save_obj(data_dict, filename)
        print( "Filename: "+filename)
    except Exception as e:
        print("Could not save file because: {}".format(e))
    return data_dict
```

```python
def plot_dynamics(data_dict, figname="PLOTS/nothing.pdf"):
    PARAMS = data_dict['PARAMS']
    # code goes here to take in and plot dynamics for wc, add and non-add
    biases = np.array([ 0*1e-3, 10*1e-3, 30*1e-3])*ev_to_inv_cm
    #biases =data_dict['biases']
    T_values =data_dict['T_values']
    #labels = ['wc shift','s', 'Redfield', 'Add. Shift', 'Add.', 'Non-add.']
    labels = [r"$\epsilon=$"+str(int(b*1000/ev_to_inv_cm))+"meV" for b in biases]
    f, ((ax1, ax2),(ax3, ax4))  = plt.subplots(2, 2, sharex=True, figsize=(14,8))
    # ['_add_shift', '_wc_shift', '_wcs']
    for i, rho_nonadd in enumerate(data_dict['rho_nonadd_list']):
        op_list = data_dict['op_list'][i]
        #GS pop
        ax1.plot(T_values, calculate_expval(rho_nonadd, op_list, obs='OO'), 
                 label=labels[i])
        ax1.set_ylabel(r"$\rho_{g}^{ss}$")
        ax1.legend()
        #DS pop
        ax2.plot(T_values, calculate_expval(rho_nonadd, op_list, obs='dark'), 
                 label=labels[i])
        ax2.set_ylabel(r"$\rho_{-}^{ss}$")
        #ax2.legend()
        
        ax3.plot(T_values, calculate_expval(rho_nonadd, op_list, obs='bright'), 
                 label=labels[i])
        ax3.set_xlabel(r"$T_1$, $T_2$ $(K)$")
        ax3.set_ylabel(r"$\rho_{+}^{ss}$")
        ax3.set_xlim(T_values[0], T_values[-1])
        #ax3.legend()
        
        ax4.plot(T_values, np.abs(np.array(calculate_expval(rho_nonadd, op_list, obs='sigma_x'))), label=labels[i])
        ax4.set_xlabel(r"$\alpha_1$, $\alpha_2$ $(cm^{-1})$")
        ax4.set_ylabel(r"$|\rho_{12}^{ss}|$")
        ax4.legend()
        ax4.set_xlabel(r"$T_1$, $T_2$ $(K)$")
        
        #ax4.set_ylabel(r"$\Re \langle\{+\rangle\!\langle -|\rangle$")
        ax4.set_xlim(T_values[0], T_values[-1])
        ax4.legend()
    plt.subplots_adjust(left=0.1, bottom=0.1, right=1, top=0.9, wspace=0.23, hspace=0.08)
    plt.savefig(figname)
    print(figname)
```

```python
plot_dynamics(data_dict_1, figname="PLOTS/PAPER/T-dependence--.pdf")
```

```python
def B_for_min_decay(PARAMS):
    Delta = PARAMS['V']*2
    eta = sqrt(Delta**2 + PARAMS['bias']**2)
    fp = sqrt(eta+PARAMS['bias'])
    fm = sqrt(eta-PARAMS['bias'])
    return (Delta/(2*PARAMS['bias']))*((4*PARAMS['bias']**2)/(Delta**2*PARAMS['mu']))*((fp+fm)/(fp-fm))
print(B_for_min_decay(PARAMS))
```
